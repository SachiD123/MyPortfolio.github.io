##### Exploratory Data Analysis on Global super 2016 ####

```python
#Import libraries
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Get data
filename = r'C:\Users\shachini dinushika\Documents\Data\PowerBIdata\global_superstore_2016.xlsx'
df = pd.ExcelFile(filename)

###load data sets 
```

```python
orders = pd.read_excel(df, "Orders")
people = pd.read_excel(df, "People")
returns = pd.read_excel(df, "Returns")

```

```python
print(orders.head(10))
print(people.head())
print(returns.head())
```

###Data cleaning and preprocessing####


```python
##check for null values
print(orders.isnull().sum())
 
print("#############################")
    
print(people.isnull().sum().sum())
    
    
print("#############################")
    
print(returns.isnull().sum().sum())
```

```python
### find information on data classes###
print(orders.info())
print(people.info())
print(returns.info())
```

```python
###find for duplicates###
print(orders.duplicated().sum())
print(people.duplicated().sum())
print(returns.duplicated().sum())
```

```python
####join data sets for EDA###
sales = orders.merge(people, on='Region', how='left')
sales = sales.merge(returns, on='Order ID', how='left')
```

###Exploratory Data Analysis###

```python
### data distribution###

sales.describe()
```

```python
sales[["Sales","Quantity","Discount","Profit","Shipping Cost"]].hist(bins=10, figsize=(20,15))
```

```python
sales[["Sales","Quantity","Discount","Profit","Shipping Cost"]].boxplot(figsize=(20,15))
```

###Finding the existence of correlations between numerical variables###

```python
#sales[["Sales"]].corr()
sales.corr(numeric_only=True)["Sales"].sort_values()
```

###visualise using scatter plots####

```python
fig, axs = plt.subplots(ncols=4,figsize=(10, 6))
sns.scatterplot(x = sales["Quantity"], y = sales["Sales"],hue = sales["Market"],data = sales, ax = axs[0])
sns.scatterplot(x = sales["Discount"], y = sales["Sales"],hue = sales["Market"],data = sales, ax = axs[1])
sns.scatterplot(x = sales["Profit"], y = sales["Sales"],hue = sales["Market"], data = sales, ax = axs[2])
sns.scatterplot(x = sales["Shipping Cost"], y = sales["Sales"],hue = sales["Market"], data = df, ax = axs[3])

  
```
